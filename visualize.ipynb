```python
import torch
import matplotlib.pyplot as plt
```

```python
model_list = ['./model_0.pt', './model_1.pt', './model_2.pt']

for model in model_list:
    checkpoint = torch.load(model)
    plt.plot(checkpoint['history']['acc'][0:100], marker='')
    print(checkpoint['duration'], checkpoint['history']['acc'][-1])
plt.legend(['No distill','Full','Masked'])
plt.show()
```
